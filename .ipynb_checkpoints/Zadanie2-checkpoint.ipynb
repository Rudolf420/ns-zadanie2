```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import LeakyReLU
import wandb
from wandb.keras import WandbCallback
import pandas as pd
```

```python
imgSize = (48,48)
batch = 32
```

```python
trainData = ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.3,
    height_shift_range = 0.3,
    horizontal_flip = False,
    zoom_range = 0.3
)

testData = ImageDataGenerator()
```

```python
trainGenerator = trainData.flow_from_directory(directory='./data/train', target_size=imgSize, batch_size=batch, class_mode='categorical')
valGenerator = testData.flow_from_directory(directory='./data/test', target_size=imgSize, batch_size=batch, class_mode='categorical')
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation=tf.keras.activations.linear, input_shape=(48,48,1)))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation=tf.keras.activations.linear))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation=tf.keras.activations.linear))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation=tf.keras.activations.linear))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64,activation=tf.keras.activations.relu))
model.add(Dense(7, activation=tf.keras.activations.softmax))
```

```python
precision = tf.keras.metrics.Precision()
recall = tf.keras.metrics.Recall()
model.compile(loss=tf.keras.losses.categorical_crossentropy, optimizer=Adam(), metrics=['accuracy', 'mse', precision, recall])
```

```python
history = model.fit(
    trainGenerator,
    steps_per_epoch=len(trainGenerator),
    epochs=1000,
    validation_data=valGenerator,
    validation_steps=len(valGenerator)
)
```
